```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.0,2.0])
y = np.array([300.0,500.00])

print(f'X train = {x}')
print(f'Y train = {y}')
```

```python
m = x.shape[0]
print('number of training example =',m)
```

```python
m = len(x)
print(m)
```

```python
# to represent each row
i = 0 #assume row 0 . in python index start with 0

x_i = x[i]
y_i = y[i]

print(f'x^({i}), y^({i}) = {x_i},{y_i}')
```

```python
plt.scatter(x,y, marker = '*',color ='r')
plt.title('price prediction')
plt.xlabel('size')
plt.ylabel('price')
plt.show()
```

```python
w = 200
b = 100
```

```python
def compute_model_output(w,b,x):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i]+b;
    return f_wb
```

```python
tmp_f_wb = compute_model_output(w,b,x)
plt.plot(x,tmp_f_wb, c ='r', label = 'prediction best fit line')
plt.scatter(x,y,marker = '*',c='b',label = 'actual value')
plt.legend()
plt.show()
```
