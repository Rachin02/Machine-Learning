```python
import numpy as np
import copy, math
import matplotlib.pyplot as plt
```

# Dataset

```python
data = np.loadtxt('houses.txt',delimiter = ',',skiprows= 1)
x_train = data[:,:4]
y_train = data[:,4]
X_features = ['size(sqft)','bedrooms','floors','age']
```

### feature scaling

```python
def zscore_normalization(X):
    mean = np.mean(X, axis = 0)
    sigma = np.std(X,axis = 0)
    x_norm = (X-mean)/sigma
    
    return x_norm
```

### compute cost

```python
def compute_cost(x,y,w,b):
    
    m = x.shape[0]
    total_cost = 0.0

    for i in range(m):
        fwb_i = np.dot(x[i],w)+b
        total_cost += (fwb_i - y[i])**2

    total_cost = total_cost/(2*m)
    
    return total_cost
```

### compute gradient

```python
def compute_gradient(x, y, w, b):
    
    m,n = x.shape
    dj_w = np.zeros(n)
    dj_b = 0

    for i in range(m):
        err = (np.dot(x[i],w)+b) - y[i]
        
        for j in range(n):
            dj_w[j] += (err*x[i,j])
            
        dj_b += err

    dj_w = dj_w/m
    dj_b = dj_b/m

    return dj_w, dj_b
    
```

### compute gradient Descent

```python
def compute_gradient_descent(x, y, w_in, b_in, alpha_, iter_):

    w = copy.deepcopy(w_in)
    b = b_in
    j_history = []
    

    for i in range(iter_):
        dw,db = compute_gradient(x, y, w, b)

        w = w - alpha_*dw
        b = b - alpha_*db

        if i<10000:
            j_history.append(compute_cost(x,y,w,b))

        if i% math.ceil(iter_/10)==0:
            print(f'iteration {i} cost: {j_history[-1]:8.2f}')

    return w, b
```

### model run to get w, b value

```python
alpha = 1.0e-1
itr = 1000
w_init = np.zeros(4)
b_init = 0

x_norm = zscore_normalization(x_train)

w_,b_ = compute_gradient_descent(x_norm, y_train, w_init, b_init, alpha, itr)
print(f'b = {b_:.2f}')
print(f'w = {w_}')
```

### prediction

```python
x_test = x_norm[:4]

print(np.dot(x_test,w_)+b_)
```
